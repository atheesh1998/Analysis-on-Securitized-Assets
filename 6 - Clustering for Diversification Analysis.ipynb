Converting multiple csv files into one dataframe.

```python
# get data file names
path = r'D:/DS/aspiration.ai/MOD6'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
final_df = pd.concat(dfs, ignore_index=True)
```

```python
new = final_df[['Symbol', 'Close Price']]
```

```python
new['returns'] = new['Close Price'].pct_change()
```

```python
new['daily_mean'] = new['returns'].mean()
new['daily_std'] = new['returns'].std()
```

```python
new['annual_mean'] = round(new['daily_mean'] * 252,2)
new['annual_std'] = round(new['daily_std'] * np.sqrt(252),2)
```

```python
new = new.dropna()
```
