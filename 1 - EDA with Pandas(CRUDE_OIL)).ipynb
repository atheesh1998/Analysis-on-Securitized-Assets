# Import essential packages and datasets

```python
import pandas as pd
import numpy as np
crude = pd.read_csv(r'CRUDE_OIL.csv')
crude.head()
```

##### Stripping the K off Column 'Vol.' inorder to perfrom operations

```python
#Remove null values to avoid future errors
crude.dropna(inplace = True)
```

```python
#initialize start, stop and step variables
start, stop, step = 0, -1, 1

# converting to string data type 
crude["Vol."]= crude["Vol."].astype(str) 
crude.head()
```

```python
crude['Newvol'] = crude['Vol.'].str.slice(start, stop, step)
crude.head()
```

```python
#Converting the Newvol Colun back to float
crude["Newvol"] = crude["Newvol"].astype(float)
crude.head()
```

##### Stripping the comma in 'Price' , 'Open', 'High', 'Low', inorder to perform operations.


```python
crude['Price'] = crude['Price'].str.replace(',', '')
crude['Open'] = crude['Open'].str.replace(',', '')
crude['High'] = crude['High'].str.replace(',', '')
crude['Low'] = crude['Low'].str.replace(',', '')
crude.head()
```

```python
#Converting the above columns to float datatype
crude["Price"] = crude["Price"].astype(float)
crude["Open"] = crude["Open"].astype(float)
crude["High"] = crude["High"].astype(float)
crude["Low"] = crude["Low"].astype(float)
crude.head()
```

```python
crude['Total'] = crude['Price'] * crude['Newvol']
crude.head()
```
